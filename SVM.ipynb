```python
# Read CSV file
import pandas as pd
training_set=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PED/PED_Training_Dataset.csv");
print("Training Set");
print(training_set);

# SVM - Training with training data set
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(training_set[["MLH","MRH","NLDP","NRDP"]],training_set["TARGET"]) 

print("SVM - Training Successfully Completed...")
```

```python
# Test SVM classifier with one new data
#newData=[0,7,0,5];  # Normal
#newData=[1,29,1,1]; # Normal
newData=[15,6,11,6]; # PED
PED_prediction=clf.predict([newData]);
print("New Data",end=" ");
print(newData);
print("Predicted Output: ",PED_prediction);
if (PED_prediction[0]==0):
    print("Healthy Macula...");
elif(PED_prediction[0]==1):
    print("PED Found...");
```
