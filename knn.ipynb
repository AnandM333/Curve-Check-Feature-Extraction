```python
# Read CSV file
import pandas as pd
training_set=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PED/PED_Training_Dataset.csv");
print("Training Set");
print(training_set);

# Train with training data set using KNN
from sklearn.neighbors import KNeighborsClassifier  
knn_classifier = KNeighborsClassifier(n_neighbors=5)  
knn_classifier.fit(training_set[["MLH","MRH","NLDP","NRDP"]],training_set["TARGET"]) 

print("KNN - Training Successfully Completed...")
```

```python
# KNN classifier with one new data
newData=[0,7,0,5];  # Normal
#newData=[1,29,1,1]; # Normal
#newData=[15,6,11,6]; # PED
KNN_PED_Prediction=knn_classifier.predict([newData]);
print("New Data",end=" ");
print(newData);
print("Predicted Output: ",KNN_PED_Prediction);
if (KNN_PED_Prediction[0]==0):
    print("Healthy Macula...");
elif(KNN_PED_Prediction[0]==1):
    print("PED Found...");
```
