```python
# To read CSV file
import pandas as pd
training_set=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PED/PED_Training_Dataset.csv");
print("Training Set");
print(training_set);

# Train with training data set using Logistic Regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(training_set[["MLH","MRH","NLDP","NRDP"]],training_set["TARGET"])

print("Logistic Regression - Training Successfully Completed...")
```

```python
# Test the LR classifier with one new data
newData=[0,7,0,5];  # Normal
#newData=[1,29,1,1]; # Normal
#newData=[15,6,11,6]; # PED
LR_PED_Prediction=model.predict([newData]);
print("New Data",end=" ");
print(newData);
print("Predicted Output: ",LR_PED_Prediction);
if (LR_PED_Prediction[0]==0):
    print("Healthy Macula...");
elif(LR_PED_Prediction[0]==1):
    print("PED Found...");
```
